```python
df = pd.read_csv("order_products__train_chunk.csv")
```

```python
arr=[]
transaction=[]
num=0
check=0
for i in df.add_to_cart_order:
    if i>check:
        arr.append(df['product_id'].iloc[num])
        num=num+1
        check=i
    else:
        transaction.append(arr)
        print("*** array ***")
        print(arr)
        arr=[]
        check=1
        arr.append(df['product_id'].iloc[num])
        num=num+1
print(transaction)
```

```python
pattern = fp.find_frequent_patterns(transaction,2)
pattern
```

```python
rules = fp.generate_association_rules(pattern,.7)
rules
```
